```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
from train_model_v2_CNN_outputs import get_ouputs_cnn
```

```python
def get_outputs(modelname, epoch):
    with open(f'./models/{modelname}/history_{epoch}.pkl', 'rb') as f:
        history = pickle.load(f)
    return history['outputs']
```

```python
with open('data_arrays_CNN.pkl', 'rb') as f:
    data_splits = pickle.load(f)

targets = data_splits['val_validRS_values']

# Rearrange to put first the 0s and then the 1s
targets = np.concatenate((targets[0::2], targets[1::2]))
```
