```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
modelname = 'cnn_v0.1'
epoch = 40
```

```python
with open('data_arrays_CNN.pkl', 'rb') as f:
    data_splits = pickle.load(f)

targets = data_splits['val_validRS_values']

outputs = np.load(f'outputs/{modelname}/epoch_{epoch}.npy')
```

```python
binary_outputs = np.where(outputs > np.mean(outputs), 1, 0)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(binary_outputs, label='Outputs')
plt.scatter(range(len(targets)), targets, label='Targets', marker='o')
plt.title('Model Outputs and Targets')
plt.xlim(0, 10)
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.legend()
plt.show()
```
