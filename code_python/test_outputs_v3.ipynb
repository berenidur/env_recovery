```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
modelname = 'autoencoder_breast_v0.1'
```

```python
# Plot the first element of predictions and actuals
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.plot(predictions[:20, 0, 0], label='Predictions')
plt.plot(actuals[:20, 0, 0], label='Actuals')
plt.title('R')
plt.legend()

# Plot the second element of predictions and actuals
plt.subplot(1, 2, 2)
plt.plot(predictions[:20, 0, 1], label='Predictions')
plt.plot(actuals[:20, 0, 1], label='Actuals')
plt.title('S')
plt.legend()

plt.show()
```
