```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
modelname = 'autoencoder_breast_v0.1'
predictions = np.load(f'models/{modelname}/validation_predictions.npy')
actuals = np.load(f'models/{modelname}/validation_actuals.npy')
```

```python
# Plot the first element of predictions and actuals
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.scatter(range(20), predictions[:20,0], label='Predictions')
plt.scatter(range(20), actuals[:20,0], label='Actuals')
plt.title('R')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.legend()

# Plot the second element of predictions and actuals
plt.subplot(1, 2, 2)
plt.scatter(range(20), predictions[:20,1], label='Predictions')
plt.scatter(range(20), actuals[:20,1], label='Actuals')
plt.title('S')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.legend()

plt.show()
```

```python
modelname = 'autoencoder_breast_v0.2' # epoch 1200
predictions = np.load(f'models/{modelname}/validation_predictions.npy')
actuals = np.load(f'models/{modelname}/validation_actuals.npy')
```

```python
# Plot R
plt.figure(figsize=(20, 5))

plt.subplot(1, 4, 1)
plt.scatter(range(20), predictions[:20,0], label='Predictions')
plt.scatter(range(20), actuals[:20,0], label='Actuals')
plt.title('R')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.legend()

# Plot S
plt.subplot(1, 4, 2)
plt.scatter(range(20), predictions[:20,1], label='Predictions')
plt.scatter(range(20), actuals[:20,1], label='Actuals')
plt.title('S')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.legend()

# Plot beta
plt.subplot(1, 4, 3)
plt.scatter(range(20), predictions[:20,2], label='Predictions')
plt.scatter(range(20), actuals[:20,2], label='Actuals')
plt.axhline(y=0.25, color='r', linestyle='--')
plt.title('beta')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.legend()

# Plot k
plt.subplot(1, 4, 4)
plt.scatter(range(20), predictions[:20,3], label='Predictions')
plt.scatter(range(20), actuals[:20,3], label='Actuals')
plt.axhline(y=1, color='r', linestyle='--')
plt.title('k')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.minorticks_on()
plt.legend()

plt.show()
```
